```python
from pathlib import Path
import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

project_root = Path('..').resolve()
print('Project root:', project_root)

src_path = project_root / 'src'
if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

from uais.data.load_cyber_data import load_cyber_data
from uais.features.cyber_features import build_cyber_feature_table
```

```python
df_cyber_raw = load_cyber_data()
print('Raw cyber shape:', df_cyber_raw.shape)
df_cyber_raw.head()
```

```python
df_cyber_raw.info()
```

```python
target_col = 'label'
print('Target distribution (0=normal,1=attack):')
print(df_cyber_raw[target_col].value_counts())
print('Attack ratio:', df_cyber_raw[target_col].mean())
```

```python
numeric_cols = df_cyber_raw.select_dtypes(include=[np.number]).columns
print('Number of numeric columns:', len(numeric_cols))
df_cyber_raw[numeric_cols].describe().T.head(20)
```

```python
drop_cols = [c for c in ['id', 'attack_cat'] if c in df_cyber_raw.columns]

df_cyber_feats = build_cyber_feature_table(
    df_raw=df_cyber_raw,
    target_column=target_col,
    drop_columns=drop_cols,
)

processed_dir = project_root / 'data' / 'processed' / 'cyber'
processed_dir.mkdir(parents=True, exist_ok=True)

cyber_feats_path = processed_dir / 'unsw_nb15_features.parquet'
df_cyber_feats.to_parquet(cyber_feats_path)
print('Saved cyber feature table to:', cyber_feats_path)
```

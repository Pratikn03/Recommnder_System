# Reports & Plots Preview
Quickly view summary tables (`reports/metrics_<domain>.csv`) and plots (`figures/reports/<domain>_roc_auc.png`).
Run after you generate reports via `python -m src.uais.reporting.make_tables` and `python -m src.uais.reporting.make_plots`.

```python
# Path setup
import sys
from pathlib import Path
project_root = Path('..').resolve()
sys.path.append(str(project_root / 'src'))
reports_dir = project_root / 'reports'
fig_dir = project_root / 'figures' / 'reports'
domains = ['fraud','cyber','behavior','nlp','vision','fusion']
print('Project root:', project_root)
```

## View tables

```python
import pandas as pd
for d in domains:
    path = reports_dir / f'metrics_{d}.csv'
    print(f"\n== {d.upper()} table ==")
    if path.exists():
        display(pd.read_csv(path))
    else:
        print('missing; run make_tables')
```

## View plots

```python
from IPython.display import Image, display
for d in domains:
    fig = fig_dir / f'{d}_roc_auc.png'
    print(f"\n== {d.upper()} plot ==")
    if fig.exists():
        display(Image(filename=str(fig)))
    else:
        print('missing; run make_plots')
```
